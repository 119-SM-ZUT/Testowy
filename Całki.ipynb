# Ca≈Çkowanie numeryczne

```python
import numpy as np
import sympy as sp
import scipy 
from scipy import integrate
# import mpmath
import matplotlib.pyplot as plt
import matplotlib as mpl
import sympy as sp
sp.init_printing()
```

```python
def f(x):
    return np.exp(-x**2)
```

```python
val, err = integrate.quad (f, -1, 1); val, err
```

```python
integrate.quad (f, -sp.oo, sp.oo)
```

```python
integrate.quad (f, -np.inf, np.inf)
```

```python
def g(x):
    return 1/np.sqrt(abs(x))
```

Funkcja $g(x)\colon\frac{1}{\sqrt{|x|}}$

```python
integrate.quad(g, -1, 1)
```

```python
x = np.linspace(-5,5, 1000)
```

```python
plt.plot(x, g(x), x , f(x));
```

```python
integrate.quad(g, -1, 1, points=[0])
```

```python
integrate.quadrature(f, -1, 1)
```

```python
integrate.quad (f, -1, 1)
```

```python
integrate.fixed_quad (f, -1, 1)
```

# Ca≈Çkowanie symboliczne

```python
x = sp.symbols("x")
```

$\int\limits_0^1 e^x dx$

```python
sp.integrate(sp.E**x, (x,0,1))
```

```python
sp.integrate(1/sp.sqrt(abs(x)), (x,-1,1))
```

```python
sp.integrate(2*sp.sqrt(1-x**2), (x, -1 , 1))
```

```python
sp.integrate(1/(x**3-1))
```

```python
sp.integrate(1/(x**3-1), (x,3,5))
```

```python
sp.factor(_)
```

```python
sp.N(_,100)
```
